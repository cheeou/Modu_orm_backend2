```python
# 리스트 기초
# 숫자형 / 문자형 / 리스트 안에 리스트(차원)
x = [1, 2, 3] #CSV 파일과 비슷 ,
y = ['apple', 'banana', 'cherry', 'melon']

print(len(x))
print(len(y))
```

```python
# 리스트의 연산 - 덧셈

a = [1, 2, 3, 4, 5]
b = [6, 7, 8]
# [1, 2, 3, 4, 5][6, 7, 8]
print(b + a) # 출력: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
# 리스트의 연산 - 곱셈
x = [1, 2]
print(x * 3) # 출력: [1, 1, 1]
```

```python
l = [1, 2, 3, 4, 5]
print(l[:])     # 출력: [1, 2, 3, 4, 5]
print(l[::-1])  # 출력: [5, 4, 3, 2, 1]
print(l[::2])   # 출력: [1, 3, 5]
print(l[0:4][::-1]) # 출력: [3, 2, 1]
```

```python
# 학생들의 총점을 구하는 코드를 작성하세요.
student_score = {
		'홍의': 97,
		'원희': 60,
		'동해': 77,
		'변수': 79,
		'창현': 89,
}
print(student_score)
type(student_score)

# 학생들의 평균 점수를 구하는 코드를 작성하세요.
# 호준: 98 점수가 추가되었습니다. 이 점수를 반영하여 학생들의 평균 점수를 구해주세요.
```

```python
# 학생들의 총 점수를 구하는 코드
score_all = student_score.values()
```

```python
total_score = sum(score_all)
```
