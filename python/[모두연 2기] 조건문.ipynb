```python
# 2시 12분 까지!!
# 카페 소개 print("")
print("안녕하세요~~ 카페 입니다.")
# 입력값 1. 현재 가지고 있는 돈 입력
input_money = int(input("현재 돈을 입력해주세요: "))
# 입력값 2. 커피 선택
input_coffee = int(input("커피를 선택해주세요: 1. 아메리카노 2. 라떼 3. 카페모카"))

# 조건 1. 커피 3종류 중 1개 선택
if input_coffee == 1:
    print("아메리카노를 선택하셨습니다.")
    price = 600
elif input_coffee == 2:
    print("라떼를 선택하셨습니다.")
    price = 1000
elif input_coffee == 3:
    print("카페모카를 선택하셨습니다.")
    price = 1500
else:
    print("잘못된 입력입니다.")
    price = 0

# 조건 2. 구매를 했을때 거스름돈 계산
if input_money >= price:
    change = input_money - price
    print(f"거스름돈은 {change}원 입니다.")
else:
    print("돈이 부족합니다.")
```

```python
def login(username, password):
    correct_username = "id123"
    correct_password = "password123"

    if username == correct_username and password == correct_password:
        return "일치 로그인 성공"
    elif username != correct_username and password == correct_password:
        return "패스워드만 맞습니다. 아이디를 확인해주세요"
    elif username == correct_username and password != correct_password:
        return "아이디만 맞고, 패스워드를 확인햊쉐요."
    else:
        return "모두 올바르지 않습니다."

# 함수 호출 예시
id = input("아이디를 입력해주세요: ")
password = input("비밀번호를 입력해주세요: ")
result = login(id, password)
print(result)
```

```python
import re

def check_password_strength(password):
    strength = 0
    suggestions = []

# 길이 검사
    if len(password) >= 12:
        strength += 2
    elif len(password) >= 8:
        strength += 1
    else:
        suggestions.append("길이를 확인하세요. 비밀번호는 최소 8자 이상입니다.")

# 대문자 검사
    if re.search(r'[A-Z]', password):
        strength += 1
    else:
        suggestions.append("대문자를 포함하세요.")


# 숫자 검사
    if any(char.isdigit() for char in password):
        strength += 1
    else:
        suggestions.append("숫자를 포함하세요.")

# 소문자 검사
    if re.search(r'[a-z]', password):
        strength += 1
    else:
        suggestions.append("소문자를 포함하세요.")

# 특수문자 검사
    if re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
        strength += 1
    else:
        suggestions.append("특수문자를 포함하세요.")

# 강도 평가
    if strength == 6:
        return "매우 강함", suggestions
    elif strength == 5:
        return "강함", suggestions
    elif strength == 4:
        return "보통", suggestions
    elif strength == 3:
        return "약함", suggestions
    else:
        return "매우 약함", suggestions

# 메인 프로그램
password = input("비밀번호를 입력하세요: ")
strength, suggestions = check_password_strength(password)
print(f"비밀번호 강도: {strength}")
print("개선 제안:", ", ".join(suggestions) if suggestions else "없음")
```
