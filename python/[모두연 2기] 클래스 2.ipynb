```python
# 1. 동물들에 대한 이름, 동물의 타입, 배고픔, 행복도
# 다마코치 프로그램!

class VirtualPet: # 명사의 첫글자는 대문자, 글자는 붙여서
    def __init__(self, name, animal_type): # __init__ 생성자 # 바깥에서 입력받을 내용들
        self.name = name
        self.animal_type = animal_type
        self.hunger = 50
        self.happiness = 50

    # 먹이주기 메서드
    def feed(self):
        # 1. 메서드에 self를 입력하면, __init__에서 설정한 인스턴스 객체를 활용할수 있다.
        self.hunger -= 10 # 먹이를 주면 배고픔이 감소함니다.
        self.happiness += 10 # 먹이를 주면 행복도가 증가합니다.
        print(f'{self.name}에게 먹이를 주었습니다.')

    def status(self):
        return f"{self.name}의 상태 : 배고픔 {self.hunger}, 행복도 {self.happiness} 입니다."

my_pet = VirtualPet("꽃님이", "말티즈")
print(my_pet.status())
my_pet.feed()
print(my_pet.status())
```

# 퀴즈 게임. 생성자와 캡슐화!

```python
class QuizGame: # 퀴즈 게임 만들기!!
    # 문제와 답이 있고, 미리 정해진 답을 유저가 맞추면 득점 / 못맞추면 실점/점수부여 x
    def __init__(self):
       # 초기에 기본 세팅이 될수 있도록 인스턴스 화를 하지 않음! 외부에서 정보를 받아오지 않음
       self._questions = {
           "내 이름은?": "JinhwanKim",
           "파이썬은?": "재밋다.",
       }
       self._score = 0

    def play(self):
        for question, answer in self._questions.items(): # items -> 딕셔너리의 키/값
            # 1. 이미 만들어진 질문을 불러오기
            user_answer = input(question + " ")
            # 2. 입력이 된값과 실제 답에 대해서 비교
            if user_answer.lower() == answer.lower():
                print("정답")
                self._score += 1
            else:
                print("오답")

        print(f"점수 : {self._score} 입니다.")

    def get_score(self):
        return self._score

    def add_question(self, question, answer):
        # 1. 질문이 해당 질문에 없으면! 추가해라
        if question not in self._questions:
            self._questions[question] = answer
            print("질문이 추가되었습니다.")
        else:
            print("이미 존재하는 질문입니다.")

game = QuizGame()
game.add_question("내 이름은?", "JinhwanKim")
game.add_question("파이썬은????", "재밋다")
game.play()
```

```python
class Character:
    def __init__(self, name, health, attack): # __init__ 생성자 # 바깥에서 입력받을 내용들 이름, 체력, 공격
        self.name = name
        self.health = health
        self.attack = attack

    def attack_enemy(self, enemy): # 이 부분은 메서드 오버라이드를 통해 구현 (틀만 만들어 놓음)
        pass

    def show_info(self): # 이름, 체력, 공격력에 대한 출력!
        print(f"이름: {self.name}, 체력: {self.health}, 공격력: {self.attack}")

class Mage(Character):
    #1. 캐릭터의 정보 + 여러분들만의 캐릭터 정보 마법사 - 지혜 / 전사 - 힘 / 지구력
    def __init__(self, name, health, attack, magic_power):
        super().__init__(name, health, attack)
        self.magic_power = magic_power

    def show_info(self): # 이름, 체력, 공격력에 대한 출력!
        print(f"이름: {self.name}, 지혜 : {self.magic_power} ,체력: {self.health}, 공격력: {self.attack}")


new_Jhkim = Mage("마법사", 100, 20, 10)
new_Jhkim.show_info()
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self , name):
        self.name = name

    @abstractmethod
    def make_sound(self):
        pass

    def introduce(self):
        print(f"저는 {self.name}입니다.")
        self.make_sound()
```

```python
class Dog(Animal):
    def make_sound(self):
        print("멍멍")

poppy = Dog("바둑이")
poppy.introduce()
```

```python
class Example:
    def __init__(self):
        self.public = "누구나 접근 가능"
        self._protected = "클래스 내부나 하위 클래스에서 사용 권장"
        self.__private = "클래스 내부에서만 사용 권장"

    def get_private(self):
        return self.__private

e = Example()
print(e.public)  # 정상 작동
print(e._protected)  # 경고는 없지만, 직접 접근은 권장되지 않음
print(e.get_private())  # 정상 작동, 클래스 메서드를 통한 접근
print(e._Example__private)  # 네임 맹글링을 통한 접근 (권장되지 않음)
```
