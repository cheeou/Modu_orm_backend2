```python
import requests
from bs4 import BeautifulSoup

response = requests.get('http://www.paullab.co.kr/stock.html')

response.encoding = 'utf-8' # 인코딩 설정
html = response.text

soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
soup.find_all('h2')[0:3]
```

```python
soup.find_all('table', class_=['table','table-hover'] )[0] # class_ : 예약어
# 예약어: 특정한 기능을 수행하도록 미리 예약되어 있는 것
```

```python
soup.select('.table > tbody > tr')[2]
```

```python
oneStep = soup.select('.main')[2] # 제주코딩베이스캠프 연구원
oneStep
```

```python
twoStep = oneStep.select('tbody > tr')[1:]
twoStep
```

```python
twoStep.select('td')[0].text # 날짜
```

```python
twoStep
```

```python
날짜 = []
종가 = []

for i in twoStep:
    날짜.append(i.select('td')[0].text)
    종가.append(int(i.select('td')[1].text.replace(',', '')))
```

```python
print(날짜, 종가)
```

```python
# 시각화
# 날짜별로 가격 변동 추이
import plotly.express as px

fig = px.line(x=날짜, y=종가, title='jejucodingcamp')
fig.show()
```
