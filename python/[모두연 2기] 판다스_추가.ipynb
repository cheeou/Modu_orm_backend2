# Value Count

```python
import pandas as pd

s = pd.Series(['A', 'B', 'C' , 'C' , 'C' , 'C' ,'C', 'B' , 'B' , 'B'])

count = s.value_counts()
print(count)
```

```python
# - `normalize`: 빈도 대신 비율을 반환합니다 (기본값: False).
# - `sort`: 결과를 정렬합니다 (기본값: True).
# - `ascending`: 오름차순으로 정렬합니다 (기본값: False).
# - `dropna`: NaN 값을 제외합니다 (기본값: True).
```

```python
# 비율로 표시하는 normalize
count = s.value_counts(normalize = True)
print(count)
```

```python
# 비율로 표시하는 normalize
count = s.value_counts(ascending = True)
print(count)
```

```python
# NaN 값 포함하기
import pandas as pd

s = pd.Series(['A', 'B', 'C' , 'C' , None, None, None, 'B' , 'B' , 'B'])
count = s.value_counts()
print(count)
```

```python
# dropna -> null값을 drop 시켜라 -> 제거하라
count_with_nan = s.value_counts(dropna = False)
print(count_with_nan)
```

```python
import numpy as np

df = pd.DataFrame({
    '값': [1, 2, 3, 4, 5],
    '함수': ['제곱', '세제곱', '제곱근', '로그', '사인']
})

def 동적_함수(row):
    if row['함수'] == '제곱':
        return row['값'] ** 2
    elif row['함수'] == '세제곱':
        return row['값'] ** 3
    elif row['함수'] == '제곱근':
        return np.sqrt(row['값'])
    elif row['함수'] == '로그':
        return np.log(row['값'])
    elif row['함수'] == '사인':
        return np.sin(row['값'])

df['결과'] = df.apply(동적_함수, axis=1)
print(df)
```

```python
print(동적_함수(df))
```

```python
import pandas as pd

log_df = pd.read_csv('web_server_log_final.csv')
user_df = pd.read_csv('user_info.csv')

print(log_df.head())
print(user_df.head())
print(log_df.describe())
```

```python
# 데이터 병합을 하시는데요.
# ip_address를 기준으로 merge left_join
merge_df = pd.merge(log_df, user_df, left_on='ip_address', right_on='ip_address', how='left')
```

```python
print(merge_df)
```

```python
### 2.2 사용자 유형별 접속 횟수(Value count, group by 사용)

### 설명:

# 각 사용자 유형(regular, premium 등)별로 접속 횟수를 계산합니다.

### 가이드라인:
# value count
1. 'user_type' 열의 값을 기준으로 그룹화합니다. (Value counts)
2. 각 그룹의 크기를 계산합니다.
# group by
3. group by를 사용하여 평균 접속 횟수를 출력합니다. (group by)
```

```python
user_type_counts = merge_df['user_type'].value_counts(normalize=True)
print(user_type_counts)

user_type_mean = merge_df.groupby('user_type')['ip_address'].count().mean()
print(user_type_mean)
```
