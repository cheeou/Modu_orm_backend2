````python
# prompt: 위의 조건문에 elif를 활용한 예제로 변경해줘

import pandas as pd

# 데이터셋 불러오기
data = pd.read_csv("example_dataset.csv")

# 각 컬럼에 대한 평균 계산
average_age = data['age'].mean()
average_height = data['height'].mean()
average_weight = data['weight'].mean()

print("Average Age:", average_age)
print("Average Height:", average_height)
print("Average Weight:", average_weight)

# If the average age is greater than 30, print "The average age is above 30."
# If the average age is between 20 and 30, print "The average age is between 20 and 30."
# Otherwise, print "The average age is below 20."
if average_age > 30:
  print("The average age is above 30.")
elif average_age > 20:
  print("The average age is between 20 and 30.")
else:
  print("The average age is below 20.")
````
